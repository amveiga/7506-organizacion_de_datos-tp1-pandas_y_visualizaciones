```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_labels = pd.read_csv('datasets/Dengue/dengue_labels_train.csv')
df_features = pd.read_csv('datasets/Dengue/dengue_features_train.csv')
```

```python
df_labels
```

```python
df_features.tail()
```

```python
def week_to_season(weekofyear):
    seasons = []
    for i in weekofyear:
        if i >= 48 or i < 9:
            seasons.append('SUM')
        elif 9 <= i and i < 22:
            seasons.append('AUT')
        elif 22 <= i and i < 35:
            seasons.append('WIN')
        elif 35 <= i and i < 48:
            seasons.append('SPR')
    return pd.Series(seasons)
```

```python
df_labels['season'] = week_to_season(df_labels['weekofyear'])
df_labels
```

```python
df_features['season'] = week_to_season(df_features['weekofyear'])
df_features['season']
```

```python
sns.set_theme(font_scale=1.5)
sns.set_style("whitegrid")
f,axs=plt.subplots(nrows=2,figsize=(15,15))
g1=sns.barplot(x='season',y='total_cases',hue='city',data=df_labels,ax=axs[0],
              palette='Reds')
g1.set_ylabel('Total Cases')
g1.set_xlabel('Seasons')
g2=sns.violinplot(x='season',y='station_avg_temp_c',hue='city',data=df_features,
                  split=True,ax=axs[1],
                  palette='flare')
g2.set_ylabel('Average Temperature')
g2.set_xlabel('Seasons')
g1.set_title('Season Comparison Between Total Cases and Average Temperature ')
```

```python
plt.subplots(figsize=(20,10))
g = sns.histplot(df_features.reanalysis_relative_humidity_percent,stat='frequency',binrange=(65,100),color='b',element='step',kde=True)
g.set_xlabel('Mean Relative Humidity')
g.set_title('Frequency of Mean Relative Humidity',fontdict={'fontsize':25})
```
