```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
soja_df = pd.read_csv('data-segunda_parte/datasetRofex4 - datasetRofex2.csv')
```

```python
soja_df
```

```python
soja_df['Fecha']
```

```python
soja_df.loc[:,'Fecha'] = pd.to_datetime(soja_df['Fecha'])
```

```python
soja_df.info()
```

```python
soja_df['Año'] = soja_df['Fecha'].dt.year
```

```python
soja_df.head()
```

```python
soja_short = soja_df[['Año','Fecha','Cierre','Aj.Dif.']]
soja_short.head()
```

```python
soja_short = soja_short.loc[soja_short['Año'] == 2019]
soja_short['Mes'] = soja_short['Fecha'].dt.month
soja_short = soja_short.sort_values(by='Mes')
soja_short
```

```python
soja_short['Aumento por Día'] = soja_short['Aj.Dif.'] > 0
soja_short
```

```python
sns.set(rc={'figure.figsize':(20,10)},font_scale=1.5)
sns.set_style('white')
ax1 = sns.boxplot(x='Mes',y='Cierre',hue= 'Aumento por Día',data=soja_short, palette={'r','b'})
#ax1 = sns.boxplot(x='Mes',y='Cierre',data=soja_short, palette={'r','b'})
ax1.set_title('Cierre de la soja en 2019',fontdict={'fontsize':20})
sns.despine(offset=10, trim=True)
#plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

Segunda visualizacion

```python
soja_short2 = soja_df[['Año','Cierre']]
precios_max = soja_short2.groupby('Año')['Cierre'].agg('max')
precios_min = soja_short2.groupby('Año')['Cierre'].agg('min')
```

```python
precios_max.head()
```

```python
dfBar = pd.DataFrame([precios_max,precios_min]).transpose()
```

```python
dfBar.head()
```

```python
dfBar.columns=['Cierre Max Año','Cierre Min Año']
```

```python
dfBar.reset_index()
```

```python
dfBar.plot(kind='bar',rot=45,title='Cierres máximos y mínimos anuales',color=(['b','r']))
ax=plt.gca()
ax.set_ylabel=('Precios')
ax.set_xlabel=('Años')
```
