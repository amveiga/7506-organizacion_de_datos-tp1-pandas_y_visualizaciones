```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
soja_df = pd.read_csv('datasets/Comportamiento de la soja/datasetRofex4 - datasetRofex2.csv')
```

```python
soja_df
```

Chequeo que todos los datos se hayan cargado correctamente.

```python
soja_df.info()
```

Casteo el tipo de dato para poder trabajar mejor.

```python
soja_df.loc[:,'Fecha'] = pd.to_datetime(soja_df['Fecha'])
```

```python
soja_df.info()
```

```python
soja_df['Año'] = soja_df['Fecha'].dt.year
```

```python
soja_df.head()
```

```python
soja_boxplot = soja_df[['Año','Fecha','Cierre','Aj.Dif.']]
soja_boxplot.head()
```

Mi idea es visualizar lo que paso en el 2019

```python
soja_boxplot = soja_boxplot.loc[soja_boxplot['Año'] == 2019]
soja_boxplot['Mes'] = soja_boxplot['Fecha'].dt.month
soja_boxplot = soja_boxplot.sort_values(by='Mes')
soja_boxplot
```

```python
soja_boxplot['Aumento Diario'] = soja_boxplot['Aj.Dif.'] > 0
soja_boxplot
```

Ploteo

```python
sns.set(rc={'figure.figsize':(20,10)},font_scale=1.5)
sns.set_style('white')
ax1 = sns.boxplot(x='Mes',y='Cierre',hue= 'Aumento Diario',data=soja_boxplot, palette={'r','g'})
ax1.set_title('Cierre de la soja en 2019',fontdict={'fontsize':20})
sns.despine(offset=10, trim=True)
```

Segunda visualizacion

```python
soja_barplot = soja_df[['Año','Cierre']]
precios_max = soja_barplot.groupby('Año')['Cierre'].agg('max')
precios_min = soja_barplot.groupby('Año')['Cierre'].agg('min')
```

```python
precios_max.head()
```

```python
dfBar = pd.DataFrame([precios_max,precios_min]).transpose()
```

```python
dfBar.head()
```

```python
dfBar.columns=['Cierre Max Año','Cierre Min Año']
```

```python
dfBar.reset_index()
```

```python
dfBar.plot(kind='bar',rot=45,title='Cierres máximos y mínimos anuales',color=(['b','r']))
ax=plt.gca()
ax.set_ylabel=('Precios')
ax.set_xlabel=('Años')
```
