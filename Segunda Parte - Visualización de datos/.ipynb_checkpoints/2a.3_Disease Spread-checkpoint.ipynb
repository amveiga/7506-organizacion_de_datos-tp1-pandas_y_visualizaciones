```python
def week_to_season(weekofyear):
    seasons = []
    for i in weekofyear:
        if i >= 48 or i < 9:
            seasons.append('SUM')
        elif 9 <= i and i < 22:
            seasons.append('AUT')
        elif 22 <= i and i < 35:
            seasons.append('WIN')
        elif 35 <= i and i < 48:
            seasons.append('SPR')
    return pd.Series(seasons)
```

```python
df_labels['season'] = week_to_season(df_labels['weekofyear'])
df_labels
```

```python
df_features['season'] = week_to_season(df_features['weekofyear'])
df_features['season']
```

```python
sns.violinplot(x='season',y='station_avg_temp_c',hue='city',data=df_features,
              split=True)
```
