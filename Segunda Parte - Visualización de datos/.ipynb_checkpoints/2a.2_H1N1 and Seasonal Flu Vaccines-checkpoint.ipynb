```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/Flu/training_set_features.csv')
```

```python
df
```

```python
dfHeatmap=df[['age_group','h1n1_concern']].dropna()
dfHeatmap['h1n1_concern']=dfHeatmap['h1n1_concern'].astype(int)
dfHeatmap.columns = ['Age Group','Level of Concern']
dfHeatmap
```

```python
dfHeatmap=dfHeatmap.groupby(['Age Group','Level of Concern']).agg({'Level of Concern':['count']})
dfHeatmap.head()
```

```python
dfHeatmap=dfHeatmap.unstack()
dfHeatmap
```

```python
dfHeatmap.columns=dfHeatmap.columns.droplevel([0,1])
dfHeatmap
```

```python
sns.set_theme(style="ticks", font_scale=2)
f = plt.figure(figsize=(15,10))
f = plt.yticks(rotation=0)
ax = sns.heatmap(dfHeatmap, cmap="YlOrBr",)
ax = ax.set_title('Age Group vs. Level of Concern',fontdict={'fontsize':25})
```

```python
dfBar = df[['income_poverty','chronic_med_condition','sex']].dropna()
```

```python
dfBar = dfBar[dfBar['chronic_med_condition'] > 0]
```

```python
dfBar=dfBar.groupby(['income_poverty','sex']).agg('sum')
```

```python
dfBar = dfBar.reset_index()
dfBar
```

```python

f = plt.figure(figsize=(15,10))
g = sns.barplot(x='chronic_med_condition',y='income_poverty',hue='sex',data=dfBar,
            palette=['violet','royalblue'])
g.set_ylabel('Income Poverty')
g.set_xlabel('Chronical Medical Condition')
l = plt.legend(title='Sex')
```
