# Ejercicio 38
Obtenga la matriz de correlación para saber idiomas distintos considerando que un
usuario sabe un idioma si indicó un nivel de 1 o superior

```python
import pandas as pd
```

```python
languages = pd.read_csv('wikipedia_dataset-2c2021/languages.csv')
```

```python
languages.head()
```

```python
languages.info()
```

```python
languages['babel_lang'].hasnans
```

```python
babelLangFillna=languages['babel_lang'].fillna('NaN')
filtroBabelLangNa=babelLangFillna == 'NaN'
languagesNans=languages[filtroBabelLangNa]
languagesNans
```

```python
languages=languages.dropna()
```

```python
languages.isna().sum()
```

```python
languages = languages[languages['babel_level']!='0']
```

```python
languages
```

```python
languages['babel_level'].value_counts()
```

```python
grouped_usrs = languages.sort_values('babel_user')
```

```python
grouped_usrs
```

```python
grouped_usrs['knows_lang'] = True
```

```python
grouped_usrs.head(10)
```

```python
pivot_grouped = grouped_usrs.pivot(index='babel_user',columns='babel_lang',values='knows_lang')
pivot_grouped = pivot_grouped.fillna(False)
```

```python
pivot_grouped
```

## Resultado

```python
lang_corr = pivot_grouped.corr('pearson')
lang_corr
```
