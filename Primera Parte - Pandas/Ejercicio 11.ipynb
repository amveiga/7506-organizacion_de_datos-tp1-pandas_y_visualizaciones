Para todos los comentarios de revisión de contenido que tengan más de 20
ocurrencias realice una matriz cuyas columnas sean esos comentarios y de índice
los usuarios/ips con valores: True si ese usuario realizó ese comentario, sino False.

```python
import pandas as pd
```

```python
contents = pd.read_csv('wikipedia_dataset-2c2021/contents.csv')
```

```python
contents.head()
```

```python
allRevisors = contents['revisor_ip'].fillna(contents['revisor_username'])
allRevisors
```

```python
contents['users_or_ip'] = allRevisors
contents.head()
```

```python
contents_short = contents[['revisor_comment','users_or_ip']]
contents_short
```

```python
contents_short['ocurrences'] = contents_short.groupby(['revisor_comment']).transform('count')
```

```python
contents_short=contents_short.sort_values('ocurrences',ascending=False)
contents_short
```

```python
contents_short = contents_short[contents_short['ocurrences'] > 20]
contents_short
```

```python
contents_short['made_comment'] = True
```

```python
contents_short['revisor_comment'].value_counts()
```

```python
contents_short['revisor_comment'].hasnans
```

```python
pivot_contents = contents_short.pivot_table(index='users_or_ip',columns='revisor_comment',values='made_comment',fill_value=False)
```
